## **변수의 종류**
---

### 수량적 특성에 의한 구분
  - 질적 변수
    - 측정하면 몇 개의 범주 중 하나에 속하게 하는 변수. 분류되어 속성을 가짐
    - 예) 성별(남,녀),인종(황인종,백인종,흑인종)
    - 비서열 
    - 서열
  - 양적 변수
    - 측정하면 양 즉, 수로 표현되는 변수 가진다.
    - 연속형
    - 비연속형 : 몇 개의 정수로만 표현

### 자료의 특성에 의한 구분
  - 이산형 변수(범주형 변수)
    - 대상들이 서로 떨어진 값을 가짐
    - 질적 변수와 비연속형 양적 변수를 모두 포함
    - 명목 변수 : 대상에 이름이 붙지만 순위는 없음
    - 순위 변수 : 순위가 있는 변수
  - 연속형 변수
    - 대상들이 서로 연속된 값을 가짐
    - 간격 변수 : 절대 0과 비율이 의미가 없음, 온도.
    - 비율 변수 : 절대 0과 비율이 의미가 있다. 곱하기 가능. 키, 몸무게, 나이, 시간


## **각 변수에 따른 통계기법**
---



### 그룹(명목 변수)들 사이에 차이가 있는지 알아보기 위한 통계분석

1. 그룹이 **2개**인지 **3개 이상**인지 파악
2. 각 그룹의 자료가 **독립적**인지, **상관관계**가 있는지 파악
3. 각 그룹에 따른 종속 변수가 **범주형**인지 **연속형**인지 파악
  - 연속형 변수일 경우
    - 정규 분포인지(sample size>30, 모수적 통계기법) 비정규 분포인지(비모수적 통계기법) 파악
  - 범주형 변수일 경우
    - 명목인지 순위인지 파악

image.png


- Mann-Whitney U test / wilcoxn signed rank test / Kruskal-Wallis test / Friedman test 는 비모수 통계 검정이다.

- 예
  > 1. 모집단에서 **고혈압이 있는 사람과 없는 사람들** 간에 BMI차이가 있는지
    - independent samples t-test
  2. **고혈압이 있는 사람과 없는 사람들** 간에 우울증 척도(Severe, Moderato, Mild)에 차이가 있는지
    - Mann-Whitney U test
  3. **고혈압이 있는 사람과 없는 사람들** 간에 혈액형 분포 차이가 있는지
    - chi-squared test
  4. 모집단에서 각 사람에 대해 고혈압 치료 전 후에 혈압 수치의 차이가 있는지
    - paired samples t-test
  5. 각 사람에 대해 고혈압 치료 전 후에 우울증 척도에 차이가 있는지
    - wilcoxn signed rank test
  6. 각 사람에 대해 수은 혈압계와 정상 혈압계로 측정했을 때 고혈압 여부(고혈압, 정상혈압)에 차이가 있는지
    - McNemar test
  7. 모집단에서 각 반에 따라 키의 차이가 있는지
    - one-way ANOVA
  8. 각 반에 따라 성적의 차이가 있는지
    - kruskal-Wallis test
  9. 각 반에 따라 혈액형의 차이가 있는지
    - chi-squared test

### 각 변수간의 상관성이 있는지 알아보기 위한 통계분석

1. 각 변수가 연속형인지 범주형인지 파악
  - 연속 변수라면 정규, 비정규분포 파악
  - 범주 변수라면 명목, 순위분포 파악

image.png

- spearman's correlation, Kendall's tau는 비모수적 검정이다.
- 예
  > 1. 집단에서 육류 소비량과 대장암 발병율은 상관성이 있는가
    - Pearson's correlation / Spearman's correlation
  2. 육류 소비량과 대장암의 Stage와는 상관성이 있는가
    - Spearman's correlation
